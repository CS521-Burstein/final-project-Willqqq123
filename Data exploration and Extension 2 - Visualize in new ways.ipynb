#  Visualize in new ways 

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from statsmodels.formula.api import ols
import statsmodels.api as sm

#Score Calculating Function
from sklearn.metrics import r2_score, explained_variance_score
```

```python
housing=df
```

```python
sns.set_style('darkgrid')
plt.xlabel('Price')
plt.title('Histogram of Sale Price')
mpl.rc("figure", figsize=(15,5))
sns.distplot(housing.SalePrice,color="y",bins=100,kde=True,kde_kws={"color":"seagreen", "lw":1 })
```

```python
plt.figure(figsize=(10,5))
plt.title('Adjusted GrLivArea and SalePrice')
plt.hist(housing['GrLivArea']*100,bins=100,color='y',stacked=True)
plt.hist(housing['SalePrice'],bins=100,color='b',stacked=True)
```

```python
x=housing[housing['TotalBsmtSF']<=4000]['TotalBsmtSF']
xy=housing[housing['TotalBsmtSF']<=4000]['SalePrice']
y=housing['SalePrice']
z=housing[housing['GarageArea']<=4000]['GarageArea']
zy=housing[housing['GarageArea']<=4000]['SalePrice']
a=housing['GrLivArea']
b=housing[housing['LotFrontage']<=200]['LotFrontage']
by=housing[housing['LotFrontage']<=200]['SalePrice']
```

```python
plt.subplot(2,2,2)
plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.title('TotalBsmtSF and SalePrice')
plt.scatter(x, xy,c='b',marker= '*',alpha=0.5,cmap='viridis')
plt.colorbar()

plt.subplot(2,2,1)
plt.title('GarageArea and SalePrice')
plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.scatter(z, zy,c='pink',marker= 'v',alpha=0.5,cmap='Spectral')
plt.colorbar()


plt.subplot(2,2,3)
plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.title('GrLivArea and SalePrice')
plt.scatter(a, y,c='green',alpha=0.5,cmap='viridis')
plt.colorbar()


plt.subplot(2,2,4)
plt.rcParams.update({'figure.figsize':(10,8), 'figure.dpi':100})
plt.title('LotFrontage and SalePrice')
plt.scatter(b, by,c='r',marker= '.',alpha=0.5,cmap='viridis')
plt.colorbar()
```
